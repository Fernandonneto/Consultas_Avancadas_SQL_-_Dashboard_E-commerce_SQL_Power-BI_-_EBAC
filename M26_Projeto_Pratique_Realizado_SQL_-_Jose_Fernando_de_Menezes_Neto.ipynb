Abaixo temos a configuração do ambiente SQL:

```python
import sqlite3
import pandas as pd
```

```python
df_transacoes = pd.read_csv("TB_TRANSACOES_PROJETO_ECOMM.csv", delimiter=';')
df_clientes = pd.read_csv("TB_CLIENTES_PROJETO_ECOMM.csv", delimiter=';')
```

```python
conn = sqlite3.connect('projeto.db')
# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
# Visualização para análise da tabela "tb_transacoes"
query = """
SELECT * FROM tb_transacoes
"""
result_df = run_query(query)
print(result_df)
```

```python
# Visualização para análise da tabela "tb_clientes"
query = """
SELECT * FROM tb_clientes
"""
result_df = run_query(query)
print(result_df)
```

```python
query = """
SELECT *
FROM tb_transacoes
LEFT JOIN tb_clientes
ON tb_transacoes.ID_CLIENT = tb_clientes.Id_client
"""
result_df = run_query(query)
print(result_df)
```

Justifique a escolha do JOIN:

A escolha do **LEFT JOIN** foi realizada considerando a estrutura e o objetivo analítico das bases de dados. A tabela **tb_transacoes** possui maior volume de registros (367 linhas e 4 colunas) em comparação à tabela **tb_clientes** (175 linhas e 5 colunas), representando de forma mais detalhada as transações do e-commerce.

O foco da análise e da construção do dashboard será no desempenho de vendas, produtos e comportamento comercial, então optou-se pelo **LEFT JOIN** para garantir que todas as transações da tabela principal (tb_transacoes) fossem preservadas, incluindo apenas as informações correspondentes da tabela **tb_clientes** quando disponíveis.

Essa abordagem evita perda de registros de vendas, possibilitando a análise de clientes, e permite análises mais completas sobre produtos, volume transacional e possível captação de clientes.

Exportando o arquivo como CSV:
